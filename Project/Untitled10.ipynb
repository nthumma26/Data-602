```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv('Current_Tax_Sale_list.csv')
```

```python
#Distribution of total_tax
plt.figure(figsize=(10,5))
(sns.distplot(data['total_tax'],bins=50,kde=True,color = "green")).set(xlabel="Sales Tax", ylabel = "Count", title='Distribution of Sales Tax')
plt.show()
```

```python
#comparing the propert type by its total_tax
plt.figure(figsize=(10,4))
sns.barplot(data['ownership'],data['total_tax'])
```

```python
#visualization of correlation matrix
corrmat = data.corr()
f, ax = plt.subplots(figsize=(8, 8))
sns.heatmap(corrmat, vmax=.8, square=True);
```
